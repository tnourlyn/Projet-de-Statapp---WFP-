```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("NIGERIA_WFP_ADMIN1.csv", sep=None, engine="python")
df["time"] = pd.to_datetime(df["time"], errors="coerce", dayfirst=True)
df = df.dropna(subset=["time", "NAME_1"])
df["year"] = df["time"].dt.year

# NDVI: prends celui que tu as (vim)
df["ndvi"] = df["vim"]

# On filtre la qualité NDVI
df = df[df["viq"] >= 80].copy()   

df = df.sort_values(["NAME_1", "time"])
print(df.head(5))
```

* **SOS_date (Start of Season – date):** the calendar date when the growing season starts (NDVI rises above a threshold and stays above it consistently).
* **EOS_date (End of Season – date):** the calendar date when the growing season ends (after the peak, NDVI falls below the threshold and stays below it).
* **LOS_days (Length of Season – days):** the duration of the growing season in days = *EOS_date − SOS_date*.
* **peak_date:** the date of the maximum NDVI in the year (when vegetation is most developed/active).
* **peak_doy (Peak Day of Year):** the day-of-year (1–365/366) corresponding to `peak_date`, useful for comparing timing across years.
* **peak_ndvi:** the peak NDVI value reached at `peak_date` (magnitude of peak vegetation greenness).
* **SOS_doy (Start of Season Day of Year):** the day-of-year corresponding to `SOS_date`.
* **EOS_doy (End of Season Day of Year):** the day-of-year corresponding to `EOS_date`.
* **ndvi_mean:** the annual mean NDVI (average vegetation greenness/activity over the year).
* **ndvi_std:** the annual NDVI variability (standard deviation); higher values indicate stronger seasonality (or noisier observations).
* **n_obs:** the number of NDVI observations used for that region-year (more observations generally means more reliable estimates).
* **threshold_T:** the NDVI threshold used to define “in-season” vs “out-of-season,” computed adaptively from the annual baseline and peak.


```python
df_ndvi = df.copy(deep=True) #copier df pour faire les lags par la suite 
```

```python
def phenology_one_year(g, consec=1, q_base=0.10, frac_amp=0.15):
    g = g.sort_values("time").dropna(subset=["ndvi"])
    y = g["ndvi"].to_numpy()
    t = g["time"].to_numpy()

    if len(g) < 6:
        return None

    base = np.nanquantile(y, q_base)
    peak_idx = int(np.nanargmax(y))
    peak_ndvi = float(y[peak_idx])
    peak_date = pd.Timestamp(t[peak_idx])

    amp = peak_ndvi - base
    if (not np.isfinite(amp)) or amp <= 0:
        return None

    T = base + frac_amp * amp
    above = y >= T
    below = y < T

    def first_consecutive(mask, start=0):
        run = 0
        for i in range(start, len(mask)):
            if mask[i]:
                run += 1
                if run >= consec:
                    return i - consec + 1
            else:
                run = 0
        return None

    sos_idx = first_consecutive(above, start=0)
    if sos_idx is None:
        return None

    eos_idx = first_consecutive(below, start=peak_idx)
    if eos_idx is None:
        eos_idx = len(g) - 1  # fallback

    sos = pd.Timestamp(t[sos_idx])
    eos = pd.Timestamp(t[eos_idx])

    if eos <= sos:
        return None

    return {
        "SOS_date": sos,
        "EOS_date": eos,
        "LOS_days": int((eos - sos).days),
        "peak_date": peak_date,
        "peak_doy": int(peak_date.dayofyear),
        "peak_ndvi": peak_ndvi,
        "SOS_doy": int(sos.dayofyear),
        "EOS_doy": int(eos.dayofyear),
        "ndvi_mean": float(np.nanmean(y)),
        "ndvi_std": float(np.nanstd(y)),
        "n_obs": int(len(g)),
        "threshold_T": float(T),
    }
```

```python
rows = []
for (region, year), g in df.groupby(["NAME_1", "year"]):
    res = phenology_one_year(g)
    if res is None:
        continue
    res["NAME_1"] = region
    res["year"] = year
    rows.append(res)

pheno = pd.DataFrame(rows)
print("pheno rows:", len(pheno))
print(pheno.head())
```

```python
print(pheno[["SOS_doy","LOS_days","peak_doy","peak_ndvi"]].describe())
print(pheno.groupby("year")[["SOS_doy","LOS_days","peak_doy","peak_ndvi"]].describe())
print(pheno.groupby("NAME_1")[["SOS_doy","LOS_days","peak_doy","peak_ndvi"]].mean().sort_values("LOS_days"))
```

```python
plt.figure(figsize=(6,5))
plt.scatter(analysis["LOS_days"], analysis["inadequate"])
plt.title("Inadequate vs LOS")
plt.xlabel("LOS_days")
plt.ylabel("inadequate")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,5))
plt.scatter(analysis["peak_ndvi"], analysis["inadequate"])
plt.title("Inadequate vs Peak NDVI")
plt.xlabel("peak_ndvi")
plt.ylabel("inadequate")
plt.tight_layout()
plt.show()
```

This scatter plot compares **inadequate** (y-axis) with the **length of season (LOS_days)** (x-axis). Points form vertical bands because LOS takes a few discrete values (around 6–10 days). Overall, there is a **negative association**: higher LOS_days tends to correspond to **lower** inadequate values. Shorter seasons (LOS ≈ 7–8) show **higher and more variable** inadequate levels, while longer seasons (LOS ≈ 10) are mostly associated with **lower** inadequate values. However, the spread within each LOS value suggests that LOS alone does not fully explain inadequate outcomes and other factors likely contribute.

This scatter plot shows the relationship between **inadequate** (y-axis) and **peak NDVI** (x-axis), a proxy for maximum vegetation greenness/productivity. Overall, there is a **negative trend**: higher peak NDVI values tend to be associated with **lower** inadequate levels. At **lower peak NDVI** (~0.50–0.65), inadequate values are generally **higher and more variable**, whereas at **higher peak NDVI** (~0.75–0.82) most observations cluster at **lower** inadequate values (roughly 0.15–0.40). However, the scatter remains wide, indicating that peak NDVI alone does not fully explain inadequate outcomes and other factors likely influence the result.

```python
cols = ["inadequate","SOS_doy","LOS_days","peak_doy","peak_ndvi","ndvi_mean","ndvi_std"]
corr = analysis[cols].corr(method="spearman")

plt.figure(figsize=(7,6))
plt.imshow(corr, interpolation="nearest", cmap="RdBu_r", vmin=-1, vmax=1)  # rouge/bleu
plt.xticks(range(len(cols)), cols, rotation=45, ha="right")
plt.yticks(range(len(cols)), cols)
plt.colorbar()
plt.title("Corrélations Spearman")
plt.tight_layout()
plt.show()
```

The Spearman correlation matrix highlights a few key drivers of inadequate outcomes. The strongest relationships indicate that later season onset (higher SOS_doy) is associated with higher inadequate, while a longer growing season (higher LOS_days) is associated with lower inadequate. Vegetation intensity also matters: higher peak_ndvi and higher mean NDVI (ndvi_mean) tend to correspond to lower inadequate, suggesting that more productive/greener seasons are linked to better conditions. In contrast, higher NDVI variability (ndvi_std) is positively associated with inadequate, consistent with more unstable vegetation dynamics. Finally, peak timing (peak_doy) shows weak correlations overall, implying that when the peak occurs is less informative than how early the season starts, how long it lasts, and how strong/stable vegetation growth is.

```python
# 1) latitude par Etat
lat_by_state = df.groupby("NAME_1")["Y"].median()

# 2) seuil médian
lat_cut = lat_by_state.median()

# 3) label
north_states = set(lat_by_state[lat_by_state >= lat_cut].index)
df["zone_ns"] = np.where(df["NAME_1"].isin(north_states), "North", "South")

print("Latitude cut:", lat_cut)
print(df[["NAME_1","Y","zone_ns"]].head())
```

```python
df = df.sort_values(["NAME_1","time"])

rows = []
for (region, year), g in df.groupby(["NAME_1", "year"]):
    res = phenology_one_year(g)   # ta fonction
    if res is None:
        continue
    res["NAME_1"] = region
    res["year"] = year
    rows.append(res)

pheno = pd.DataFrame(rows)

# Ajouter la zone Nord/Sud à pheno
pheno = pheno.merge(lat_by_state.rename("lat_state"), on="NAME_1", how="left")
pheno["zone_ns"] = np.where(pheno["NAME_1"].isin(north_states), "North", "South")

# Comparaisons simples
summary = pheno.groupby("zone_ns")[["SOS_doy","EOS_doy","LOS_days","peak_doy","peak_ndvi","ndvi_mean"]].agg(["mean","median","std","count"])
print(summary)
```

```python
from scipy.stats import mannwhitneyu

def mw_test(var):
    a = pheno.loc[pheno["zone_ns"]=="North", var].dropna()
    b = pheno.loc[pheno["zone_ns"]=="South", var].dropna()
    stat, p = mannwhitneyu(a, b, alternative="two-sided")
    return pd.Series({"n_N":len(a), "n_S":len(b), "U":stat, "pvalue":p,
                      "median_N":a.median(), "median_S":b.median()})

tests = pd.concat([mw_test(v) for v in ["peak_ndvi","ndvi_mean","SOS_doy","LOS_days","peak_doy"]], axis=1).T
print(tests.sort_values("pvalue"))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for v in ["peak_ndvi","ndvi_mean","SOS_doy","LOS_days","peak_doy"]:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=pheno, x="zone_ns", y=v)
    sns.stripplot(data=pheno, x="zone_ns", y=v, color="black", alpha=0.25)
    plt.title(f"{v} : North vs South")
    plt.show()
```

The boxplot compares peak NDVI (maximum seasonal greenness) between the South and North zones. The South has a noticeably higher central tendency: its median is around 0.76–0.78, and most observations cluster roughly between 0.72 and 0.80, indicating a stronger and more consistent vegetation peak. The North shows a lower median, around 0.63–0.65, and a wider interquartile range, meaning peak greenness is more variable across observations. Several points in the North fall below 0.55, highlighting cases of particularly weak peak vegetation. Overall, the distribution suggests higher and more stable peak vegetation conditions in the South, while the North experiences lower and more heterogeneous peak NDVI.

```python
df_ndvi["year_month"] = df_ndvi["time"].dt.to_period("M").dt.to_timestamp()

panel = (
    df_ndvi.groupby(["NAME_1", "year_month"], as_index=False)
           .agg(ndvi_mean=("ndvi", "mean"),
                inadequate=("inadequate", "mean"),
                n_obs=("ndvi", "size"))
           .sort_values(["NAME_1", "year_month"])
)

panel["year"] = panel["year_month"].dt.year
panel["month"] = panel["year_month"].dt.month

print("panel rows:", len(panel), "| states:", panel["NAME_1"].nunique())
panel.head(10)
```

```python
panel["ndvi_mu_m"] = panel.groupby(["NAME_1", "month"])["ndvi_mean"].transform("mean")
panel["ndvi_sd_m"] = panel.groupby(["NAME_1", "month"])["ndvi_mean"].transform("std")
panel["ndvi_anom"] = (panel["ndvi_mean"] - panel["ndvi_mu_m"]) / panel["ndvi_sd_m"]
panel.loc[panel["ndvi_sd_m"] == 0, "ndvi_anom"] = 0

print(panel[["NAME_1","year_month","ndvi_mean","ndvi_mu_m","ndvi_sd_m","ndvi_anom"]].head(12))
print("ndvi_anom summary:")
print(panel["ndvi_anom"].describe())
```

```python
for k in [1,2,3]:
    panel[f"ndvi_anom_lag{k}"] = panel.groupby("NAME_1")["ndvi_anom"].shift(k)
    panel[f"ndvi_mean_lag{k}"] = panel.groupby("NAME_1")["ndvi_mean"].shift(k)

print(panel[["NAME_1","year_month","ndvi_anom","ndvi_anom_lag1","ndvi_anom_lag2"]].head(15))
```

```python
h = 1
panel["y_target"] = panel.groupby("NAME_1")["inadequate"].shift(-h)
model_df = panel.dropna(subset=["y_target", "ndvi_anom_lag1"])

print(panel[["NAME_1","year_month","inadequate","y_target"]].head(12))
print("share of missing y_target:", panel["y_target"].isna().mean())
```

```python
model_df = panel.dropna(subset=["y_target","ndvi_anom_lag1"])
print("model_df rows:", len(model_df))
model_df.head(10)
```

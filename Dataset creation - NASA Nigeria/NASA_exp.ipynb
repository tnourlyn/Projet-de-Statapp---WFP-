```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Assure-toi que geodata_NASA_Na est un GeoDataFrame
gdf = gpd.GeoDataFrame(geodata_NASA_Na, geometry='geometry', crs="EPSG:4326")

# Créer une colonne pour savoir si le point n'a pas d'Admin2
gdf['no_admin2'] = gdf['NAME_2'].isna()

# Plot
fig, ax = plt.subplots(figsize=(10, 10))

# Tracer les polygones Admin2 du Nigeria
admin2_gdf.boundary.plot(ax=ax, color='black', linewidth=1)

# Tracer les points avec Admin2
gdf[~gdf['no_admin2']].plot(ax=ax, markersize=5, color='blue', label='Avec Admin2')

# Tracer les points sans Admin2
gdf[gdf['no_admin2']].plot(ax=ax, markersize=5, color='red', label='Sans Admin2')

plt.legend()
plt.title("Points météo : avec et sans Admin2")
plt.show()
```

We note most pixels for which NAME_2 is Na are outside Nigeria, let's create another data frame only containing the values inside Nigeria. We also do not need the columns giving information pertaining to snow levels as all the values are null. We can also remove some columns from the admin_2 data set  

```python
gdf_Na = geodata_NASA_Na.dropna(subset=['NAME_2'])
gdf_Na = gdf_Na.drop(columns = ['SnowCover_inst','SnowDepth_inst', 'Snowf_tavg', 'NL_NAME_1', 'VARNAME_2', 'NL_NAME_2', 'CC_2', 'HASC_2'])
```

For the remaining NaN we can use the forward fill method

```python
print(gdf_Na.columns)
```

```python
gdf_Na['time'] = pd.to_datetime(gdf_Na['time'])
```

```python
#Appliquons la méthode forward fill qui prend la dernière valeur connue de la variable pour remplacer un Na (sachant qu'on en a seuelement 1200)
gdf_Na = gdf_Na.sort_values(by=['NAME_2', 'time']) #On trie par zone et date 
gdf_Na.head(100)
```

We can see that we have successfully sorted the data frame, allowing us to use the forward fill method

```python
#Maintenant qu'on a trié chronologiquement on groupe les données par NAME_2 pour éviter de remplacer les Na par des données appartenant à un admin 2 différent
gdf_Na = gdf_Na.groupby('NAME_2').ffill()
```

```python
import geopandas as gpd

gdf_Na = gpd.GeoDataFrame(
    geodata_NASA_Na[geodata_NASA_Na['NAME_2'].notna()],  # garder toutes les colonnes utiles
    geometry='geometry',
    crs=geodata_NASA_Na.crs  # même système de coordonnées
)
```

```python
admin2_mean = (
    gdf_Na
    .groupby('NAME_2')[varNames]
    .mean()
    .reset_index()
)
admin2_mean.head()
```

```python
admin2_map = admin2_gdf.merge(admin2_mean, on='NAME_2', how='left')
```

3. Seasonal  analysis 

```python
gdf_Na['time'].dtype
```

```python
gdf_Na['month'] = gdf_Na['time'].dt.month
gdf_Na['year'] = gdf_Na['time'].dt.year
```

```python
monthly_nigeria = (
    gdf_Na
    .groupby('month')['Rainf_f_tavg']
    .mean()
)

monthly_nigeria.plot(figsize=(8,5))
plt.title("Saisonnalité des précipitations – Nigeria")
plt.xlabel("Mois")
plt.ylabel("Pluie moyenne")
plt.show()
```

```python
admin2_stress = (
    gdf_Na
    .groupby('NAME_2')[['Rainf_f_tavg', 'Tair_f_tavg']]
    .mean()
    .reset_index()
)

admin2_stress['hydric_stress'] = (
    admin2_stress['Tair_f_tavg'] /
    admin2_stress['Rainf_f_tavg']
)
```

```python
stress_map = admin2_gdf.merge(admin2_stress, on='NAME_2')

stress_map.plot(
    column='hydric_stress',
    cmap='OrRd',
    legend=True,
    figsize=(10,8)
)
plt.title("Indice de stress hydrique – Nigeria (Admin2)")
plt.axis('off')
plt.show()
```
